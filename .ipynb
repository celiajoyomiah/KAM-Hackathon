```python
import pandas as pd
```

```python
# Function to load and examine the data

def load_and_examine_data(file_path):
    try:
        # Load the data from the specified file path
        data = pd.read_csv(file_path)

        # Display the shape, columns and the first few rows of the dataset
        print("----------------------------Details about the data----------------------------\n")
        print("----------------------------Shape of the dataset----------------------------\n")
        display(data.shape)
        print()
        print("----------------------------Columns of the dataset----------------------------\n")
        display(data.columns)
        print()
        print("----------------------------Head of the dataset----------------------------\n")
        display(data.head())
        print("----------------------------Tail of the dataset----------------------------\n")
        display(data.tail())
        print()
        # Display information about the dataset

        print("\n----------------------------Data information ----------------------------\n")
        display(data.info())
        print("\n----------------------------Descriptive Statistics of the dataset ----------------------------\n")
        display(data.describe())

        return data

    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Replace with your data file path
file_path = "data.csv" 
data = load_and_examine_data(file_path)
```

## Description of columns
1. destination_code: A code representing the destination country or region where the goods are being sent.

2. destination_id: A unique identifier for the destination, possibly used internally to reference specific destinations in the database.

3. destination_name: The name of the destination country or region.

4. discharge_port: The port where the goods will be unloaded or discharged upon reaching the destination.

5. entry_number: A unique number assigned to the shipment or entry for tracking purposes.

6. entry_status: The current status of the shipment or entry, indicating whether it is pending, cleared, in transit, etc.

7. excise_duty: The amount of excise duty levied on the goods, typically a tax imposed on specific goods produced or sold within a country.

8. export_duty: The amount of duty levied on goods being exported out of the country.

9. hscode_code: The Harmonized System (HS) code assigned to the product, which classifies goods for customs purposes.

10. hscode_description: A description of the product based on its HS code, providing details about the type of goods being imported or exported.

11. import_duty: The amount of duty levied on goods being imported into the country.

12. import_vat: The Value Added Tax (VAT) applied to imported goods.

13. origin_code: A code representing the origin country or region from where the goods are being shipped.

14. origin_id: A unique identifier for the origin, possibly used internally to reference specific origins in the database.
15. origin_name: The name of the country or region where the goods were produced or originated.

16. product_id: A unique identifier for the product, used within databases to track and manage specific products.

17. product_name: The name of the product being imported, exported, or otherwise recorded.

18. quantity: The amount or volume of the goods being handled, usually measured in units, kilograms, liters, etc.

19. railway_development_levy: A specific levy or tax imposed on goods to fund railway infrastructure development, typically applied during importation.

20. reg_date: The registration date of the entry, indicating when the goods were recorded or registered in the system.

```python
# Creating a function that returns null, duplicated and placeholder values in the dataset.
def data_prep(df):
    print('-------------------------Missing Values Check---------------------------------------\n')
    print(f'Number of null values in each column in the dataset:\n{df.isnull().sum()}\n')
    print('-------------------------Duplicated Values Check------------------------------------\n')
    print(f'Number of duplicated values in the dataset: {df.duplicated().sum()}\n')
    print('-------------------------Placeholder Values Check-----------------------------------\n')
    for column in df.columns:
        unique_values = df[column].unique()
        placeholders = [value for value in unique_values if str(value).strip().lower() in ['placeholder', 'na', 'n/a', '?']]
        placeholder_count = len(placeholders)
    
        print(f"Column: '{column}'")
        print(f"Placeholders found: {placeholders}")
        print(f"Count of placeholders: {placeholder_count}\n")
        
# Checking in our dataset.
data_prep(data)
```
